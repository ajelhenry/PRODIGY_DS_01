```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_final = pd.read_csv('./task01_Dataset.csv',  encoding='utf-8')
df_final.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


chosen_country = 'Belgium'

country_data = df_final[df_final['Country Name'] == chosen_country]

numeric_columns = country_data.select_dtypes(include='number').columns
years = pd.to_numeric(numeric_columns.str.replace(r'\D+', '', regex=True))  # Explicitly set regex=True

# Extract population data for the chosen country
population = country_data[numeric_columns].values.flatten()

# Create a figure with subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot for population over the years
axs[0].bar(years, population, color='lightgreen', edgecolor='black')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Population')
axs[0].set_title(f'Population of {chosen_country} Over the Years')

# Histogram of population data
axs[1].hist(population, bins=20, color='lightgreen', edgecolor='black')
axs[1].set_xlabel('Population')
axs[1].set_ylabel('Frequency')
axs[1].set_title(f'Distribution of Population in {chosen_country}')

plt.tight_layout()
plt.show()
```

The left subplot displays the population of the chosen country over the years using a bar plot, while the right subplot presents the distribution of the population data using a histogram.

```python
import matplotlib.pyplot as plt

chosen_country = 'United States'
country_data = df_final[df_final['Country Name'] == chosen_country]

years = country_data.columns[2:]
years_numeric = [int(year) for year in years if year.isdigit()]
years_str = [str(year) for year in years_numeric]  # Convert to strings
population = country_data[years_str].values.flatten()

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(years_str, population, color='skyblue', label='Population')
ax1.set_xlabel('Year')
ax1.set_ylabel('Population', color='skyblue')
ax1.set_title(f'Population Trend and Distribution of {chosen_country}')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Rotate x-axis labels for better visibility
ax1.set_xticks(years_str)  # Set x-ticks to the actual years
ax1.set_xticklabels(years_str, rotation=45)  

# Histogram - Distribution of Population
ax2 = ax1.twinx()
ax2.hist(population, bins=20, color='lightgreen', edgecolor='black', alpha=0.7, label='Population Distribution')
ax2.set_ylabel('Frequency', color='lightgreen')
ax2.tick_params(axis='y', labelcolor='lightgreen')

fig.legend(loc='upper left')

plt.show()
```

In this visualization, the bar chart represents the population trend over the years for the chosen country, the United States, while the histogram displays the distribution of population data. By combining both plots, we can gain insights into both the overall trend and the distribution of population data in a single visualization.

### Bar Chart (Population over the Years):


The bar chart shows the population trend over the years. Each bar represents the population count for a specific year, with the x-axis representing the years and the y-axis representing the population count. The bars are colored sky blue for better visibility.

### Histogram (Distribution of Population):


The histogram displays the distribution of population data. It shows how frequently different population ranges occur within the dataset. The x-axis represents different population ranges, divided into bins, while the y-axis represents the frequency or count of data points within each bin. In this visualization, 20 bins are used. The histogram bars are colored light green with black edges.

### Dual-Axis Plot:


The bar chart and histogram are combined into a single visualization using a dual-axis plot. Both plots share the same x-axis (years) but have different y-axes. The left y-axis corresponds to the population count in the bar chart, while the right y-axis corresponds to the frequency of population ranges in the histogram. Legends on the upper left corner indicate which line corresponds to which type of data (population trend or population distribution)

### Overall Interpretation:


Examining the combined visualization provides insights into how the population of the United States has changed over the years. The bar chart reveals overall population trends, such as periods of growth or decline, while the histogram complements this by showing the distribution of population data, highlighting where population concentrations lie. Together, these visualizations offer a comprehensive understanding of both the population trend and its distribution for the chosen country.
